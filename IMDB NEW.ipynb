```python
## Profit of a movie 

# Budget and Revenue

# Profit - R-B

# Apply a function that returns R-B

```

```python
data['revenue']-data['budget']
```

```python
def profit_func(df):
    return df['revenue']-df['budget']
```

```python
data['profit']=data.apply(profit_func,axis=1)
```

```python
data.head(5)
```

```python
## Encode the gender variable

# male : -9999/0/m
# female: -8888/1/f
```

```python
def gender_encoding(G):
    if G=="Male":
        return 'MMMM'
    if G=="Female":
        return 'FFFFFF'
    else:
        return 'Null Value'
```

```python
data['GenderEncoding']=data['gender'].apply(gender_encoding)
```
